```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10, 15))

ax1.bar(range(1, 9), range(1, 9), color='red', edgecolor='black', hatch="/")
ax1.bar(range(1, 5), [6] * 4, bottom=range(1, 5),
        color='blue', edgecolor='black', hatch='//')
ax1.set_xticks(np.arange(0, 9))

bars = ax2.bar(range(1, 9), range(1, 9), color='yellow', ecolor='black') + \
    ax2.bar(range(1, 5), [6] * 4, bottom=range(1, 5),
            color='green', ecolor='black')
ax2.set_xticks(np.arange(0, 9))

ax3.bar(range(1, 9), range(1, 9), color='darkblue', ecolor='black') + \
    ax3.bar(range(1, 5), [6] * 4, bottom=range(1, 5),
            color='green', ecolor='black')
ax3.set_xticks(np.arange(0, 9))

plt.show()
```
