```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

np.random.seed(0)
X = np.array([1, 1, -1, -1, 2]).reshape(-1, 1)
Y = np.array([1, -1, 1, -1, 0])

clf = svm.LinearSVC(max_iter = 10000)
clf.fit(X, Y)
slope = clf.coef_
intercept = clf.intercept_
print(slope)
print(intercept)
```

### Нарисую гиперплоскости

```python
x = np.linspace(-5,5,10)


# setting the axes at the centre
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, -0.21*x - 0.105, 'b', label='y = -0.21*x - 0.105')
plt.plot(x, 0.59 * x - 0.96, 'c', label='y = 0.59 * x - 0.96')
plt.plot(x, -0.210533*x - 0.1052735, 'r', label='y=-0.210533*x - 0.1052735')
plt.plot([1], [1], marker='o', markersize=10, color="red")
plt.plot([1], [-1], marker='o', markersize=10, color="red")
plt.plot([-1], [1], marker='x', markersize=10, color="blue")
plt.plot([-1], [-1], marker='x', markersize=10, color="blue")
plt.plot([2], [0], marker='x', markersize=10, color="blue")

plt.legend(loc='upper left')

# show the plot
plt.show()
```
